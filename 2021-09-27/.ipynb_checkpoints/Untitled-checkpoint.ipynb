```python
ewallet = tabela[(tabela['Forma de Pagamento']=='Ewallet')]['Preco'].sum()
credicard = tabela[(tabela['Forma de Pagamento']=='Credit card')]['Preco'].sum()
cash = tabela[(tabela['Forma de Pagamento']=='Cash')]['Preco'].sum()
```

```python
listaLegenda=['Ewallet', 'Dinheiro', 'Cartão de Crédito']
listaSomatorio=[ewallet, cash, credicard]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(listaLegenda, listaSomatorio, color='green')
plt.xlabel('Forma de Pagamento')
plt.ylabel('Valor Pago')
plt.title('Comparação entre as formas de pagamentos')
plt.show()
```

```python
tabela['Cidade'].unique()
```

```python
tb_yangon = tabela[(tabela['Cidade']=='Yangon')]
```

```python
tb_yangon
```

```python
tb_yangon=tb_yangon.reset_index(drop=True)
```

```python
tb_yangon
```

```python
tb_yangon['Hora'].min()
```

```python
tb_yangon['Hora'].max()
```

```python
for x in range(len(tb_yangon)):
    nova_hora=str(tb_yangon['Hora'][x][:2])+'-'+str(int(tb_yangon['Hora'][x][:2])+1)
    tb_yangon['Hora'][x]=nova_hora
```

```python
lista_horas = tb_yangon['Hora'].unique()
lista_medias=list()
```

```python
for hora in lista_horas:
    lista_medias.append(tb_yangon[(tb_yangon['Hora']==hora) & (tb_yangon['Forma de Pagamento']=='Cash')]['Total'].sum())
```

```python
plt.bar(lista_horas, lista_medias, color='green')
plt.xlabel('Período')
plt.ylabel('Soma do valor')
plt.title('Soma do Valor por Período')
plt.show()
```
